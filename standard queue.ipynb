```python
class Queue:
    CAPACITY = 20
    def __init__(self):
        self.data = [None]*Queue.CAPACITY
        self.size = 0
        self.front = 0
        
    def len(self):
        return self.size
    
    def empty(self):
        return self.size == 0
    
    def first(self):
        if self.empty():
            return ("Queue is empty")
        return self.data[self.front]
    
    def deque(self):
        if self.empty():
            return ('Queue is empty')
        answer = self.data[self.front]
        self.data[self.front] = None
        self.front = (self.front + 1)%len(self.data)
        self.size -= 1
        return answer
    
    def resize(self, cap):
        old = self.data
        #allocate list with new capacity
        self.data = [None]*cap
        walk = self.front
        #only consider existing elements
        for k in range(self.size):
            #intentionally shift indices
            self.data[k] = old[walk]
            walk = (1 + walk)%len(old)
            #use old size as modulus
            #front has been aligned
            self.front = 0
            
    def enqueue(self, e):
        if self.size == len(self.data):
            self.resize(2*len(self.data))
        avail = (self.front + self.size)%len(self.data)
        self.data[avail] = e
        self.size += 1
     
    def show(self):
        print(self.data)
        
q = Queue()
q.enqueue(9)
q.show()
q.len()
q.enqueue('onew')
q.enqueue(3)
q.enqueue('sew')
q.enqueue(90)
q.enqueue('hey')
q.enqueue(809)
q.enqueue(60)
q.enqueue(30)
q.show()
print(q.len())
q.deque()
q.deque()
q.show()
q.len()
print(q.first())
print(q.empty())
q.resize(30)
q.show()
print(q.len())

```

```python
print('Name: Panashe W Madzudzo')
print('Reg No: R204568J')
n = int(input('Enter the size of the array'))
queue = [0]*n
print('1. Enqueue')
print('2. Dequeue')
print('3. Display')
print('4. Exit')

def enqueue():
    if queue.count(0)==0:
        print('Queue overflow')
    else:
        top = int(input('Enter a number to push'))
        for i in range(0,n):
            if queue[i] == 0:
                queue[i]=top
                break
                
def dequeue():
    if queue.count(0) == len(queue):
        print('Queue underflow')
    else:
        for i in range(0,n):
            if queue[i] != 0:
                queue[i] = 0
                break
                
def display():
    for i in queue:
        if i != 0:
            print(i)
            
while(1):
    choice = int(input('Enter a choice'))
    if choice == 1:
        enqueue()
    if choice == 2:
        dequeue()
    if choice == 3:
        display()
    if choice == 4:
        print('Exit')
        break
    else:
        print('Invalid number')
    
                
        
```
